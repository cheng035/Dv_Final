# Crona Virus trend this year
During 2020, millions of people get infected by crona virus, there are over 60 M infected people since now and the increasing trends hasn't been slowed in the United States according to the graph. Some countries like Germany, Koera, China have slowed the increasing trends. And for most countries, they have controlled the number of new death caused by Crona Virus.

```python
@ipywidgets.interact(specific_date=date_list,name=country_list)
def specific_info(specific_date='2020-09-02',name='United States'):
    b=pd.read_csv('covid.csv')
    mask1=b['location']==name
    mask4=b['date']==specific_date
    mask_t=mask1&mask4
    single=b[mask_t]
    print(name,"'s information on",specific_date)
    print(single)

# here I build a interacitve map to show the trends in different countries
@ipywidgets.interact(style=plt.style.available,name=country_list,start_date=date_list,end_date=date_list)
def make_plot(style='seaborn_dark',start_date='2020-01-03',end_date='2020-09-02',name='United States'):
#     # generate a basic information of a country
  
    b=pd.read_csv('covid.csv')
    mask1=b['location']==name
    
    mask2=b['date']>start_date
    mask3=b['date']<end_date

    mask=mask1&mask2&mask3
    

    
    united=b[mask]
 
    
    b = b.sort_values(by = 'date')
      
    with plt.style.context(style):
        fig,ax=plt.subplots(1,1,figsize=(20,15))
        ax.plot(united["date"],united["total_cases"],label='total_cases')
        ax.plot(united["date"],united["total_deaths"],label='total_deaths')
        ax.plot(united["date"],united["new_deaths"],label='new_cases')
        ax.plot(united["date"],united["new_cases"],label='cases')
        ax.set_xlabel(name)
        ax.set_ylabel('cased')
        ax.set_yscale('log')
        plt.legend()
        plt.show()
```
